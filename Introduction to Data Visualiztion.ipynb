```python
# 0. Import the library
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

# 2. Setup Plot
fig,ax = plt.subplots(figsize=(10,6)) # figsize in inches > Hight and widht

#3. Plot Data
ax.plot(x,y)

#4. Customize Plot
ax.set_title('My_First Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.grid()  # Adding grid

# 5. Save & Display Plot
plt.savefig('my_plot.png')
plt.show()
```

## Two Options fro Subplots

```python
# Subplot Option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax1.set_title('x/2')
ax2.scatter(np.random.randn(100), np.random.randn(100));
ax2.set_title('Scatter Plot')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set_title('Nut Butter Prices on Bar Graph')
ax4.hist(np.random.randn(1000));
ax4.set_title('Histogram')
```

```python
# Subplots option 2
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to Each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randn(100), np.random.randn(100));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which Plot is suitable for me : Pyplot or Matplotlib OO method
* When plotting something quickly for analysis, pyplot is good.
* When plotting some more advance, use the OO method

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
over_50.plot.scatter(x='age',
                     y = 'chol',
                     c = 'target');
```

```python
# Using the OO Method
fig, ax = plt.subplots( figsize = (10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax = ax);
ax.set_xlim([45,100]);
```

```python
## OO Method from Scratch
# Creating a figure and axes
fig , ax = plt.subplots ( figsize = (10 , 6) )

# Scatter plot
scatter = ax.scatter ( x = over_50['age'] ,
                       y = over_50['chol'] ,
                       c = over_50['target'] )
# Customize
ax.set ( title = 'Heart Diseases vs Cholesterol Levels' ,
         xlabel = 'Age' ,  # Corrected
         ylabel = 'Cholesterol Levels' )

# Add legend
ax.legend ( *scatter.legend_elements () , title = 'Target' )

#Adding a red horizontal Line
ax.axhline (over_50['chol'].mean(), color='red', linestyle='--')

```

```python
over_50.head()
```

```python
# Sub Plot of Chol , age and Thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10, 10))
# Add data to X0
scatter = ax0.scatter(x = over_50['age'] , y = over_50['chol'] , c = over_50['target'])

# Customize

ax0.set(title ='Heart Disease vrs Cholestraol Levels',
        xlabel = 'Age',
        ylabel = 'Cholesterol Levels')
# Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add a horizoontal line
ax0.axhline(over_50['chol'].mean(), color='red', linestyle='--')



# Replicating the Same Graph with the Thalach Column
# Add data to ax1
scatter = ax1.scatter(x = over_50['age'] ,
                      y = over_50['thalach'] ,
                      c = over_50['target'])

# Customize
ax1.set(title = 'Heart disease vs max Heart Rate',
        xlabel = 'Age',
        ylabel = 'Max Heart Rate')

# Add a legend
ax1.legend(*scatter.legend_elements(), title = 'Target')

#Add a horizontal line -  Mean line
ax1.axhline(over_50['thalach'].mean(),
            color='blue',
            linestyle='--')

# Add a tittle to the title
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

## This Plot shows some presentation of Heart Data Analysis

```python
# More on the Plotting:
plt.style.use('ggplot');

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10, 10))
# Add data to X0
scatter = ax0.scatter(x = over_50['age'] ,
                      y = over_50['chol'] ,
                      c = over_50['target'],
                       cmap = 'winter')

# Customize
ax0.set(title ='Heart Disease vrs Cholestraol Levels',
        xlabel = 'Age',
        ylabel = 'Cholesterol Levels')

#Chaning the X axis limits
ax0.set_xlim(50,80)


# Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add a horizoontal line
ax0.axhline(over_50['chol'].mean(), color='red', linestyle='--')



# Replicating the Same Graph with the Thalach Column
# Add data to ax1
scatter = ax1.scatter(x = over_50['age'] ,
                      y = over_50['thalach'] ,
                      c = over_50['target'],
                       cmap = 'winter')

# Customize
ax1.set(title = 'Heart disease vs Max Heart Rate',
        xlabel = 'Age',
        ylabel = 'Max Heart Rate')

#Chaning the X axis limits
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)
# Add a legend
ax1.legend(*scatter.legend_elements(), title = 'Target')

#Add a horizontal line -  Mean line
ax1.axhline(over_50['thalach'].mean(),
            color='blue',
            linestyle='--')

# Add a tittle to the title
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
