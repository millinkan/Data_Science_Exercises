```python
# Second method (using object-oriented interface)
fig = plt.figure() # Create a figure
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
ax.plot(x, y);
```

```python
# Third method (using subplots)
fig, ax = plt.subplots()
ax.plot(x, [50,100,200,250,300]);
type(fig), type(ax)
```

```python
# 0. Import the library
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4,5]
y = [11,22,33,44,55]

# 2. Setup Plot
fig,ax = plt.subplots(figsize=(10,6)) # figsize in inches > Hight and widht

#3. Plot Data
ax.plot(x,y)

#4. Customize Plot
ax.set_title('My_First Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.grid()  # Adding grid

# 5. Save & Display Plot
plt.savefig('my_plot.png')
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the Data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Using same data to plot scatter plot
fig,ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig,ax = plt.subplots()
ax.scatter(x, np.sin(x)); # np.sin(x) returns values between -1 and 1 . Sine Wave
```

```python
# Make a plot from Dixtiona numpy array
nut_butter_prices = {'Almonds Butter': 10, 'Peanuts Butter': 8, 'Walnuts Butter': 12}
fig, ax = plt.subplots()
ax.bar(x=nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set_title('Nut Butter Prices')
ax.set_ylabel('Price ($)');
```

```python
fig,ax = plt.subplots() # Horizontal Bars
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())); # Convert X and Y into list
```

```python
# Make Some Data for Historaph
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x); # Always remember to terminate the plot with ";"
```

## Two Options fro Subplots

```python
# Subplot Option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax1.set_title('x/2')
ax2.scatter(np.random.randn(100), np.random.randn(100));
ax2.set_title('Scatter Plot')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set_title('Nut Butter Prices on Bar Graph')
ax4.hist(np.random.randn(1000));
ax4.set_title('Histogram')
```

```python
# Subplots option 2
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to Each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randn(100), np.random.randn(100));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]','', regex  = True)
#Include the statement regex=True any time reg is used in pads
car_sales
```

```python
car_sales ['Price'] = car_sales['Price'].str[:-1]
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2025', periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#Ploting the Total Sales
car_sales.plot(x='Sale Date', y = 'Total Sales');
```
